```python
import copy
```

```python
def compare(l1,l2):
    for i in l1:
        if i not in l2:
            return False
    return True
```

```python
def check(l1,l2):
    if l2 not in l1:
        return False
    return True
```

```python
def swap(a,b):
    return b,a
```

```python
def moveZeroPosition(data1,row,col):
    newData=[]
    if row>=0 and col >=0:
#         print(row,col)
        if (col+1)<=2:
            data=copy.deepcopy(data1)
            data[row][col+1],data[row][col]=swap(data[row][col+1],data[row][col])
            newData.append(data)
#             print(data1[row][col+1],data1[row][col])
#             print(data)
            
        if (row+1)<=2:
            data=copy.deepcopy(data1)
            data[row+1][col],data[row][col]=swap(data[row+1][col],data[row][col])
            newData.append(data)
#             print(data1[row][col],data1[row+1][col])
#             print(data)
            
        if (col-1)>=0:
            data=copy.deepcopy(data1)
            data[row][col-1],data[row][col]=swap(data[row][col-1],data[row][col])
            newData.append(data)
#             print(data[row][col-1],data[row][col])
#             print(data)
            
        if (row-1)>=0 :
            data=copy.deepcopy(data1)
            data[row-1][col],data[row][col]=swap(data[row-1][col],data[row][col])
            newData.append(data)
#             print(data[row-1][col],data[row][col])
#             print(data)
            
    return newData 
```

```python
def display(key,values,data,inp):
    temp=[]
    while compare(data,inp)==False:
        if data in key:
            temp.append(data)
            data=values[key.index(data)]
    temp.reverse()
    count=1
    for l in temp:
        for x in l:
            print(" "*1,end='')
            for y in x:
                print(y,end=' ')
            print(end='\n')
        print("")
        
        if count!=len(temp):
            for x in range(3):
                print(" "*(x),end="")
                print(" *"*(3-x),end="")
                print("")
            print("")
        count-=-1
```

```python
def findZeroPosition(data):
    c=0
    for i in data:
        if 0 in i:
            return c,i.index(0)
        c=c+1
```
